# Data Exploration and Visualization

*   Load Titanic Dataset
*   Analyze key statistics for each feature
*   Vusualize relationships among features






```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Loding the data and print first 5 rows
titanic_data = pd.read_csv('https://raw.githubusercontent.com/brem-21/Titanic/refs/heads/main/Data/train.csv')
titanic_data.head()
```

```python
# Information of the data
titanic_data.info()
```

```python
# Key statistics for each feature
titanic_data.describe(include='all')
```

```python
# Converting the columns to objects for a clearer overview and Easy Analysis
titanic_data['Survived'] = titanic_data['Survived'].map({0: 'No', 1: 'Yes'})

titanic_data['Pclass'] = titanic_data['Pclass'].map({1: '1st', 2: '2nd', 3: '3rd'})

titanic_data['Embarked'] = titanic_data['Embarked'].map({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})
```

```python
# Viewing the transformed data
titanic_data.head()
```

```python
# Count plot to visualize survival rate among Gender
plt.figure(figsize=(8, 6))  # Set the figure size for the plot
sns.countplot(data=titanic_data, x=titanic_data['Sex'], hue=titanic_data['Survived'], palette='coolwarm')  # Bar plot with Survived  as hue
plt.title(f'Survival By Gender')  # Set the title for each plot
plt.xlabel('Gender')  # Label for x-axis
plt.ylabel('Number of Survival')  # Label for y-axis
plt.show()
```

```python
# Count plot to visualize survival rate among age
plt.figure(figsize=(8, 6))  # Set the figure size for the plot
sns.countplot(data=titanic_data, x=titanic_data['Pclass'], hue=titanic_data['Survived'], palette='coolwarm')  # Bar plot with Survived  as hue
plt.title(f'Survival By Ticket Class')  # Set the title for each plot
plt.xlabel('Ticket Class')  # Label for x-axis
plt.ylabel('Number of Survival')  # Label for y-axis
plt.show()
```

```python
# Count plot to visualize survival rate among Gender
plt.figure(figsize=(8, 6))  # Set the figure size for the plot
sns.countplot(data=titanic_data, x=titanic_data['Embarked'], hue=titanic_data['Survived'], palette='coolwarm')  # Bar plot with Survived  as hue
plt.title(f'Survival By Port of Embarkation')  # Set the title for each plot
plt.xlabel('Port of Embarkation')  # Label for x-axis
plt.ylabel('Number of Survival')  # Label for y-axis
plt.show()
```

```python
# Count plot to visualize survival rate among Gender
plt.figure(figsize=(8, 6))  # Set the figure size for the plot
sns.countplot(data=titanic_data, x=titanic_data['Embarked'], hue=titanic_data['Pclass'], palette='coolwarm')  # Bar plot with Survived  as hue
plt.title(f'PORT OF EMBARKATION BY TICKET CLASS')  # Set the title for each plot
plt.xlabel('Port of Embarkation')  # Label for x-axis
plt.ylabel('Count of Passengers')  # Label for y-axis
plt.show()
```
